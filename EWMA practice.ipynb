```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
airline=pd.read_csv('C:\\Users\\dell\\Downloads\\UDEMY_TSA_FINAL\\UDEMY_TSA_FINAL\\Data\\airline_passengers.csv',index_col='Month')
```

```python
airline.head()
```

```python
airline.dropna(inplace=True)
```

```python
airline.index=pd.to_datetime(airline.index)
```

```python
airline.index
```

```python
airline['6monthSMA']=airline['Thousands of Passengers'].rolling(window=6).mean()
```

```python
airline['12monthSMA']=airline['Thousands of Passengers'].rolling(window=12).mean()
```

```python
airline.plot()
```

```python
airline['EWMA-12']=airline['Thousands of Passengers'].ewm(span=12).mean()
```

```python
airline[['Thousands of Passengers','EWMA-12']].plot(figsize=(10,6))
```
